完成如下功能：书店店员有一张链表形式的书单，每个节点代表一本书，节点中的值表示书的编号。

为更方便整理书架，店员需要将书单倒过来排列，就可以从最后一本书开始整理，逐一将书放回到书架上。

请倒序返回这个书单链表。

````python
from typing import Optional, List

class ListNode:
    def __init__(self, val=0, next=None, prev=None):
        self.val = val
        self.next = next
        self.prev = prev

def reverseBookList(head: Optional[ListNode]) -> ListNode:
    prev = None
    current = head

    while current is not None:
        next_node = current.next  # 暂存下一个节点
        current.next = prev       # 翻转当前节点指针
        prev = current            # 移动 prev 指针到当前节点
        current = next_node       # 移动到下一个节点

    return prev

# 构建一个链表 [3,2,6,4]
def build_linked_list(vals: List[int]) -> ListNode:
    if not vals:
        return None
    
    head = ListNode(vals[0])
    current = head
    for val in vals[1:]:
        current.next = ListNode(val)
        current.next.prev = current
        current = current.next
    
    return head

# 打印链表以便于验证结果
def print_linked_list(head: ListNode):
    current = head
    while current is not None:
        print(current.val, end=' ')
        current = current.next
    print()

# 测试
book_list = build_linked_list([3, 2, 6, 4])
print("原链表:")
print_linked_list(book_list)

reversed_book_list = reverseBookList(book_list)
print("翻转后的链表:")
print_linked_list(reversed_book_list)
````

````python
import time
import functools

def timing_decorator(verbose=False):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            start_time = time.time()
            try:
                result = func(*args, **kwargs)
            finally:
                end_time = time.time()
                elapsed_time = end_time - start_time
                if verbose:
                    print(f"Function '{func.__name__}' executed in {elapsed_time:.6f} seconds with args: {args}, kwargs: {kwargs}")
                else:
                    print(f"Function '{func.__name__}' executed in {elapsed_time:.6f} seconds")
            return result
        return wrapper
    return decorator

# 示例使用
@timing_decorator(verbose=True)
def example_function(x, y):
    time.sleep(1)  # 模拟耗时操作
    return x + y

result = example_function(5, 3)
print("Result:", result)
````

Python 的 open 函数是一个内置函数，用于打开文件并返回一个文件对象，这个文件对象可以用来进行读写操作。

open 函数是进行文件操作的基础方法，支持多种模式和选项。以下是对 open 函数的详细解释：

### 函数签名

```python
open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
```

### 参数说明

- file:

要打开的文件的路径。这可以是一个相对路径或绝对路径。

- mode:

表示文件的打开模式，是一个字符串。常见的模式包括：
'r': 只读模式（默认值）。如果文件不存在，会抛出 FileNotFoundError。
'w': 写入模式。会创建文件或截断（清空）文件。
'a': 附加模式。在文件末尾打开并写入数据，而不截断文件。
'b': 二进制模式。例如，'rb' 或 'wb' 用于二进制读取或写入。
't': 文本模式（默认）。用于文本文件的读写。
'+': 更新模式。读写模式的组合（例如，'r+' 允许同时读写）。

- buffering:

控制文件的缓冲。默认是 -1，选择系统默认的缓冲策略。
	- 0 表示无缓冲。
	- 1 表示行缓冲。
	- 大于 1 的值表示使用固定大小的缓冲区。

- encoding:

指定打开文件时使用的编码。这通常用于文本模式下读取文件，例如 'utf-8'。

- errors:

指定在编码或解码时发生错误时要采取的措施。常用值包括 'ignore', 'replace' 等。

- newline:

控制换行符的处理。可以是 None, '', '\n', '\r', or '\r\n'。

- closefd:

如果为 False，并且给 file 参数传递的是文件描述符而非文件路径，close 方法将不会关闭文件描述符。

- opener:

是一个可选的自定义打开器（callable），该调用将被用于代替 os.open 来打开文件描述符。

- 返回值

open 函数返回一个文件对象，用于文件的读写操作。根据所使用的模式，文件对象可以提供方法如 read(), write(), seek(), 和 close()。

- 示例用法
#### 打开文件进行读取
```python
with open('example.txt', 'r', encoding='utf-8') as f:
    content = f.read()
    print(content)
```

# 实现一个自定义的上下文管理器

需求概述
设计一个自定义上下文管理器类 FileManager，用于安全地打开和关闭文件资源。该上下文管理器应能够自动处理文件的打开和关闭操作以及处理异常情况。

具体要求
类的属性：

filename：要打开的文件名。
mode：文件打开模式（例如，'r' 读取模式，'w' 写入模式）。
方法实现：

__init__：初始化 FileManager 对象，接收文件名和模式。
__enter__：进入上下文管理器，打开文件，并返回文件对象。
__exit__：退出上下文管理器，确保文件被正确关闭。应处理任何由于文件操作导致的异常。
异常处理：

捕获和处理在文件操作期间可能出现的异常，并确保文件在任何情况下都能被正确关闭。

````python
class FileManager:
    def __init__(self, filename, mode):
        """
        初始化 FileManager 对象。

        参数：
        filename (str): 要打开的文件的名称。
        mode (str): 打开文件的模式，例如 'r', 'w', 'a' 等。
        """
        self.filename = filename
        self.mode = mode
        self.file = None

    def __enter__(self):
        """
        在进入上下文管理器时调用。

        尝试打开指定的文件并返回文件对象。
        如果打开文件时出现异常，打印错误信息并返回 None。

        返回：
        文件对象或 None（如果发生错误）。
        """
        try:
            self.file = open(self.filename, self.mode)
            return self.file
        except Exception as e:
            print(f"Error opening file: {e}")
            return None

    def __exit__(self, exc_type, exc_val, exc_tb):
        """
        在退出上下文管理器时调用。

        关闭文件，并处理关闭文件时可能出现的异常。

        参数说明：
        exc_type, exc_val, exc_tb: 异常的类型、值和追踪信息（如果存在）。

        返回：
        False 表示如果在上下文中有异常，它会继续传播。
        """
        if self.file:
            try:
                self.file.close()
            except Exception as e:
                print(f"Error closing file: {e}")
        # Return False to propagate exceptions if any
        return False

# 示例使用
try:
    with FileManager('example.txt', 'w') as f:
        if f:
            f.write("Hello, world!")
except Exception as e:
    print("An error occurred:", e)

# 尝试使用读取一个不存在的文件
try:
    with FileManager('non_existent.txt', 'r') as f:
        if f:
            content = f.read()
except Exception as e:
    print("An error occurred:", e)
````
